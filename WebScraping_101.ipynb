##  BeautifulSoup

####Often times, website APIs are either not available or the data we want to collect is not available from the API.
####We can use BeautifulSoup!

BeautifulSoup is an open source Web Scraping framework for Python

Beautiful Soup Constructor takes an XML or HTML documument in the form of a string.   It parses the document and creates a corresponding data structure in memory

```python
from bs4 import BeautifulSoup
import urllib2
wiki = "http://en.wikipedia.org/wiki/Machine_learning"
page = urllib2.urlopen(wiki)
soup = BeautifulSoup(page)
print soup
```

```python
print soup.title
```

```python
# Find all <ul> elements, then Find all tags with a - (ie link tags)
for headline in soup('span', {'class' : 'mw-headline'}):
    print(headline.text)
    links = headline.find_next('ul').find_all('a')
    for link in links:
        print('*', link.text)
```
